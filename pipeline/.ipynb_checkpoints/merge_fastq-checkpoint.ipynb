```python
import glob
import subprocess
import time
import os

file_paths = {
    "run1": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/2025_01_17_Hybrid_GEX_Run1_L2/",
    "run1_a": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/2025_01_17_Hybrid_GEX_Run1_L2_a/",
    "hyb_p1e2_r02": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E2_R02_20250202/",
    "run2_1": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/2025_01_30_Hybrid_GEX_L2_Run2_1/",
    "hyb_p1e2_r03": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E2_R03_20250203/",
    "hyb_p1e3_r01": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E3_R01_20250203/",
    "hyb_p1e3_r02": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E3_R02_20250206/",
    "hyb_p1e3_r03": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E3_R03_20250209/",
    "hyb_p1e3_r04": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E3_R04_20250210/",
    "hyb_p1e2_r04": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E2_R04_20250210/",
    "hyb_p1e4_r01": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E4_R01_20250211/",
    "hyb_p1e4_r02": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E4_R02_20250213/",
    "hyb_p2e1_r01": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P2E1_R01_20250213/",
    "hyb_p2e1_r04": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P2E1_R04_20250214/",
    "hyb_p1e3_r05": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E3_R05_20250217/",
    "hyb_p1e4_r03": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E4_R03_20250217/",
    "hyb_p1e4_r04": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E4_R04_20250218/",
    "hyb_p1e4_r05": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P1E4_R05_20250220/",
    "hyb_p2e2_r01": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P2E2_R01_20250222/",
    "hyb_p2e2_r02": "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/HYB02E_LIB2_GEX_P2E2_R02_20250224/",
}

outpath = "/nfs/turbo/umms-indikar/shared/projects/hybrid_reprogramming/data/fastq/gene_expression/library_2_reads_merged/"

start_time = time.time()

for sample_id, fpath in file_paths.items():
    print(f"\n=== Processing {sample_id} ===")
    sample_start = time.time()

    files = glob.glob(f"{fpath}/*.fastq.gz")
    outfile = f"{outpath}{sample_id}.fastq.gz"

    if not files:
        print(f"  WARNING: No files found. Skipping {sample_id}.")
        continue

    # Calculate total input size (in GB)
    input_size_bytes = sum(os.path.getsize(f) for f in files)
    input_size_gb = input_size_bytes / (1024 ** 3)

    print(f"  Found {len(files)} files.")
    print(f"  Search path: {fpath}")
    print(f"  Total input size: {input_size_gb:.2f} GB")
    print(f"  Saving merged output to: {outfile}")

    command = f"cat {' '.join(files)} > {outfile}"
    output = subprocess.run(command, shell=True, capture_output=True)

    if output.returncode == 0:
        print(f"  Merge successful!")
    else:
        print(f"  ERROR during merge:")
        print(output.stderr.decode())
        continue

    # Calculate output file size (in GB)
    if os.path.exists(outfile):
        output_size_bytes = os.path.getsize(outfile)
        output_size_gb = output_size_bytes / (1024 ** 3)
        print(f"  Output file size: {output_size_gb:.2f} GB")
    else:
        print(f"  WARNING: Output file not found.")

    sample_end = time.time()
    print(f"  Time for {sample_id}: {sample_end - sample_start:.2f} seconds")

    # break  # REMOVE this line if you want to process all samples

total_time = time.time() - start_time
print(f"\n*** Merging complete. Total elapsed time: {total_time:.2f} seconds ***")
```
