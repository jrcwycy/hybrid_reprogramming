```python
# Marker genes
def load_pathway(fpath):
    result = []
    with open(fpath) as f:
        for line in f:
            split_line = [x for x in line.strip().split('\t') if x]  # Remove empty strings directly

            row = {'label': split_line[0]}
            for gene in split_line[1:]:
                row[gene] = 1

            result.append(row)

    df = pd.DataFrame(result)
    df = df.fillna(0.0).set_index('label').astype(bool).T  # Chained operations for clarity

    return df

mpath = "/nfs/turbo/umms-indikar/shared/projects/RECODE/marker_genes/PanglaoDB_Augmented_2021.txt"
pang = load_pathway(mpath)

# # Get gene sets
fb_genes = pang[pang['Fibroblasts']].index
myocyte_genes = pang[pang['Myocytes']].index
myoblast_genes = pang[pang['Myoblasts']].index
myofb_genes = pang[pang['Myofibroblasts']].index
ipsc_genes = pang[pang['Pluripotent Stem Cells']].index

print(f"Number of genes for Fibroblasts: {len(fb_genes)}")
print(f"Number of genes for Myocytes: {len(myocyte_genes)}")
print(f"Number of genes for Myoblasts: {len(myoblast_genes)}")
print(f"Number of genes for Myofibroblasts: {len(myofb_genes)}")
print(f"Number of genes for iPSCs: {len(ipsc_genes)}")
```

# Pseudobulk

Aggregated raw counts by summing across 'pooled_condition', creating a 4 x 25,126 AnnData object.
Summed counts are normalized to 10,000 and log-transformed.

```python
function = 'sum'

agg = sc.get.aggregate(
    adata,
    by='pooled_condition',
    func=function,
    layer='raw_counts',
)

agg.X = agg.layers[f"{function}"].copy()

sc.pp.normalize_total(
    agg,
    target_sum=1e4,
)

agg.layers['norm'] = agg.X.copy()

sc.pp.log1p(agg)
agg.layers['log_norm'] = agg.X.copy()

agg
```

## Pairwise differences per gene

```python
layer = 'log_norm'
agg.X = agg.layers[layer].copy()
print(f"Using {layer} layer...")

agg_df = pd.DataFrame(agg.X, index=agg.obs_names, columns=agg.var_names).T
print(agg_df.shape)

agg_df.head()
```

```python
pairwise_diffs = []
for cond1, cond2 in combinations(agg_df.columns, 2):
    diff = agg_df[cond1] - agg_df[cond2]
    col_name = f"{cond1} - {cond2}"
    pairwise_diffs.append(diff.rename(col_name))

# Combine into a new DataFrame
diff_df = pd.concat(pairwise_diffs, axis=1)
diff_df.head()
```

```python
diff_df = diff_df.loc[:, diff_df.columns.str.startswith('Control')]
print(diff_df.shape)
```

```python
# Calculate the max absolute difference across all pairwise comparisons per gene
gene_variability = diff_df.abs().max(axis=1)

n_genes = 30

# Select top N most different genes
top_genes = gene_variability.nlargest(n_genes).index  # Change 30 to however many you want
diff_subset = diff_df.loc[top_genes]

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 6, 8#11

sns.heatmap(diff_subset, cmap="vlag", center=0, annot=False, cbar_kws={"shrink": 0.5})

plt.title("Expression Differences Between Conditions (Top Variable Genes)", pad=10)
plt.xlabel("")
plt.ylabel("")
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```

```python
def find_go_terms_for_genes(go_df, gene_list, filter_string=None):
    """
    Find GO terms from a GO_Biological_Process dataframe that contain any of the given genes.

    Parameters:
    - go_df: pandas DataFrame where columns are GO terms and values are gene names.
    - gene_list: list of gene names (strings) to search for.
    - filter_string: optional string to filter GO term columns.

    Prints:
    - For each gene, the list of GO terms it's associated with (filtered if specified).
    """
    # Optionally filter columns
    if filter_string:
        go_df = go_df.loc[:, go_df.columns.str.contains(filter_string, case=False)]

    # Transpose for easier lookup
    transposed = go_df.apply(lambda col: col.dropna().astype(str).tolist())

    for gene in gene_list:
        matching_terms = [term for term, genes in transposed.items() if gene in genes]
        if matching_terms:
            print(f"\nGene: {gene}")
            for term in matching_terms:
                print(f"  - {term}")
        # else:
        #     print(f"\nGene: {gene} — No matching GO terms found.")
```

```python
def get_go_terms_per_gene(go_df, gene_list, search_terms=None):
    """
    Returns a DataFrame with one column per gene, each containing the GO terms
    the gene is part of. GO terms are filtered to include any of the search terms
    in their name (case-insensitive).

    Parameters:
    - go_df: DataFrame with GO terms as columns and gene names as values.
    - gene_list: List of gene names to search for.
    - search_terms: Optional list of substrings to filter GO term column names (OR logic).

    Returns:
    - DataFrame with one column per gene; rows list GO terms that contain that gene.
    """
    # Ensure gene list is uppercase (or match whatever casing your dataset uses)
    gene_list = [gene.strip() for gene in gene_list]

    # Step 1: Filter GO term columns
    if search_terms:
        pattern = "|".join([f"{term}" for term in search_terms])
        filtered_cols = go_df.columns[go_df.columns.str.contains(pattern, case=False, regex=True)]
        filtered_df = go_df[filtered_cols]
    else:
        filtered_df = go_df

    # Step 2: Build a mapping of each gene to matching GO terms
    gene_to_terms = {gene: [] for gene in gene_list}

    # Loop through columns (GO terms)
    for go_term in filtered_df.columns:
        genes_in_term = filtered_df[go_term].dropna().astype(str).str.strip().tolist()
        for gene in gene_list:
            if gene in genes_in_term:
                gene_to_terms[gene].append(go_term)

    # Step 3: Normalize column lengths and create result DataFrame
    max_len = max((len(terms) for terms in gene_to_terms.values()), default=0)
    for gene in gene_to_terms:
        gene_to_terms[gene] += [None] * (max_len - len(gene_to_terms[gene]))

    result_df = pd.DataFrame(gene_to_terms)

    return result_df
```

```python
def find_genes_by_go_keyword(filtered_results_df, keyword):
    keyword = keyword.lower()
    matching_genes = []

    for gene in filtered_results_df.columns:
        terms = filtered_results_df[gene].dropna().astype(str)
        if any(keyword in term.lower() for term in terms):
            matching_genes.append(gene)

    print(f"Number of matching genes for {keyword}: {len(matching_genes)}")
    return matching_genes


go_muscle = find_genes_by_go_keyword(go_terms_df, 'muscle')
go_fib = find_genes_by_go_keyword(go_terms_df, 'fibroblast')
go_ecm = find_genes_by_go_keyword(go_terms_df, 'extracellular matrix')
go_stress = find_genes_by_go_keyword(go_terms_df, 'stress')
go_cc = find_genes_by_go_keyword(go_terms_df, 'cell cycle')
go_immune = find_genes_by_go_keyword(go_terms_df, 'immune')
go_dna = find_genes_by_go_keyword(go_terms_df, 'DNA')
```

```python
### Identify regions of genes that are the most different from controls -- highlight in the plot

samples = ['mmMYOD1', 'siPRRX1', 'siPRRX1/mmMYOD1']
group_size = 5
highlight_dict = {}

# Ensure 'Start' is numeric and sort cdf
cdf_sorted = cdf.copy()
cdf_sorted['Start'] = pd.to_numeric(cdf_sorted['Start'])
cdf_sorted = cdf_sorted.sort_values(by='Start').copy()

for sample in samples:
    # Compute absolute diff from Control
    diffs = (cdf_sorted[sample] - cdf_sorted['Control']).abs().values

    # Sliding window mean diffs
    from numpy.lib.stride_tricks import sliding_window_view
    windowed = sliding_window_view(diffs, group_size)
    window_means = windowed.mean(axis=1)

    # Get top window start indices
    top_idxs = np.argsort(window_means)[-10:]

    # Get gene names in those top windows
    top_genes = []
    for idx in top_idxs:
        top_genes.extend(cdf_sorted.iloc[idx:idx + group_size].index.tolist())

    # Store as a set
    highlight_dict[sample] = set(top_genes)
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = (10, 6)

color_map = ["tab:gray", "tab:orange", "tab:purple", "tab:cyan"]

fig, axs = plt.subplots(4, 1)

all_samples = ['Control'] + samples

# Collect all highlighted genes across all samples
all_highlighted_genes = set().union(*highlight_dict.values())

x = np.arange(len(cdf_sorted))  # x positions for bars

for i, sample in enumerate(all_samples):
    ax = axs[i]

    if sample == 'Control':
        # Get non-highlighted bars
        non_highlight_mask = [gene not in all_highlighted_genes for gene in cdf_sorted.index]
        highlight_mask = [gene in all_highlighted_genes for gene in cdf_sorted.index]

        # Plot non-highlighted bars first
        ax.bar(x[non_highlight_mask], cdf_sorted[sample].values[non_highlight_mask],
               color="tab:gray", width=2.0, align='center')

        # Plot highlighted bars on top
        ax.bar(x[highlight_mask], cdf_sorted[sample].values[highlight_mask],
               color='k', width=2.0, align='center')

    else:
        # Get non-highlighted bars
        non_highlight_mask = [gene not in highlight_dict[sample] for gene in cdf_sorted.index]
        highlight_mask = [gene in highlight_dict[sample] for gene in cdf_sorted.index]

        # Plot non-highlighted bars first
        ax.bar(x[non_highlight_mask], cdf_sorted[sample].values[non_highlight_mask],
               color=color_map[i], width=2.0, align='center')

        # Plot highlighted bars on top
        ax.bar(x[highlight_mask], cdf_sorted[sample].values[highlight_mask],
               color='k', width=2.0, align='center')

    ax.set_ylabel(sample)
    ax.set_xlabel("")
    ax.set_xticks([])
    ax.set_ylim([0, 5])
    ax.set_xlim([0, len(cdf_sorted)])

axs[-1].set_xlabel('Chromosome 19 (genes ordered by position)')

sns.despine()
plt.tight_layout()
plt.show()
```
